```python
import numpy as np
from pathlib import Path
import tifffile as tiff
import matplotlib.pyplot as plt

from lavian_et_al_2025.data_location import master_all
```

```python
ref_path = master_all / "s11_anatomy"
anatomy = tiff.imread(ref_path / "s11_anatomy.tif")
```

```python
fig, ax = plt.subplots(2, 4, figsize=(8, 4))
for i in range(8):
    c = np.mod(i, 4)
    r = i // 4
    
    ax[r,c].imshow(np.rot90(anatomy[(i*2)+3], 3), cmap="gray_r", vmax=0.7)
    
    if i > 0:
        ax[r,c].axis('off')
```

```python
fig.savefig(ref_path / 'anatomy_image.pdf', dpi=300)
```
