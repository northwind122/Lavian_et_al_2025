```python
master = Path(r"Z:\Hagar\confocal\240731")

file_pre_wb = str(master / "f4_pre_wb.ome.tiff")
file_pre_ipn = str(master / "f4_pre_ipn.ome.tiff")

master_post = Path(r"Z:\Hagar\confocal\240802")
file_post_wb = str(master_post / "f4 post 2day wb.ome.tiff")
file_post_ipn = str(master_post / "f4 post 2day ipn.ome.tiff")
```

```python
img_ipn = tiff.imread(file_pre_ipn)
img_wb = tiff.imread(file_pre_wb)
```

```python
img_ipn_post = tiff.imread(file_post_ipn)
img_wb_post = tiff.imread(file_post_wb)
```

```python
path = Path(r"Z:\Hagar\confocal")
tiff.imsave(path / 'f4_pre_wb_red.tiff', img_wb[1])
tiff.imsave(path / 'f4_post_wb_red.tiff', img_wb_post[1])
tiff.imsave(path / 'f4_pre_wb_green.tiff', img_wb[0])
tiff.imsave(path / 'f4_post_wb_green.tiff', img_wb_post[0])
```

```python
path = Path(r"Z:\Hagar\confocal")
path = Path(r"Z:\Hagar\confocal")
tiff.imsave(path / 'f4_pre_ipn_red.tiff', img_ipn[1])
tiff.imsave(path / 'f4_post_ipn_red.tiff', img_ipn_post[1])
tiff.imsave(path / 'f4_pre_ipn_green.tiff', img_ipn[0])
tiff.imsave(path / 'f4_post_ipn_green.tiff', img_ipn_post[0])
```

```python
red_ipn = img_ipn[1]
green_ipn = img_ipn[0]

red_wb = img_wb[1]
green_wb = img_wb[0]


red_ipn_post = img_ipn_post[1]
green_ipn_post = img_ipn_post[0]

red_wb_post = img_wb_post[1]
green_wb_post = img_wb_post[0]
```

```python
num_planes = np.shape(red_wb)[0]
red_wb_norm = np.zeros_like(red_wb, dtype=float)
for i in range(num_planes):
    tmp_plane = zscore(red_wb[i])
    red_wb_norm[i] = tmp_plane
```

```python
num_planes = np.shape(green_wb)[0]
green_wb_norm = np.zeros_like(green_wb, dtype=float)
for i in range(num_planes):
    tmp_plane = zscore(green_wb[i])
    green_wb_norm[i] = tmp_plane
```

```python
plane_s = 0
plane_e = -20
    
red_ipn_avg = np.nanmean(red_ipn[plane_s:plane_e], axis=0)
green_ipn_avg = np.nanmean(green_ipn[plane_s:plane_e], axis=0)

plane_s = 0
plane_e = 200
    
red_wb_avg = np.nanmean(red_wb[plane_s:plane_e], axis=0)
#red_wb_avg = np.nanmean(red_wb_avg, axis=2)
green_wb_avg = np.nanmean(green_wb[plane_s:plane_e], axis=0)
#green_wb_avg = np.nanmean(green_wb_avg, axis=2)
```

```python
plane_s = 0
plane_e = -1
    
red_ipn_avg_post = np.nanmean(red_ipn_post[plane_s:plane_e], axis=0)
green_ipn_avg_post = np.nanmean(green_ipn_post[plane_s:plane_e], axis=0)

plane_s = 0
plane_e = -1
    
red_wb_avg_post = np.nanmean(red_wb_post[plane_s:plane_e], axis=0)
green_wb_avg_post = np.nanmean(green_wb_post[plane_s:plane_e], axis=0)
```

```python
fig, ax = plt.subplots(2,4, figsize=(7, 5))
```

```python
for i in range(4):
    ax[0, i].axis('off')
    ax[1, i].spines['right'].set_visible(False)
    ax[1, i].spines['top'].set_visible(False)
```

```python
height, width = red_wb_avg.shape
merged_image = np.zeros((height, width, 3), dtype=np.uint8)
merged_image[:,:,0] = red_wb_avg *255 # Red channel
merged_image[:,:,1] = green_wb_avg *100 # Green channel
merged_image = np.clip(merged_image, 0, 255).astype(np.uint8)

tmp_view = np.ma.masked_where(merged_image < 200, merged_image)
ax[0, 0].imshow(tmp_view)
ax[1, 0].imshow(merged_image)

merged_image = np.zeros((height, width, 3), dtype=np.uint8)
merged_image[:,:,0] = red_wb_avg_post *255 # Red channel
merged_image[:,:,1] = green_wb_avg_post *100 # Green channel
merged_image = np.clip(merged_image, 0, 255).astype(np.uint8)

tmp_view = np.ma.masked_where(merged_image < 200, merged_image)
ax[0, 1].imshow(tmp_view)
ax[1, 1].imshow(merged_image)


height, width = red_ipn_avg.shape
merged_image = np.zeros((height, width, 3))
merged_image[:,:,0] = red_ipn_avg *1.5 # - 50
merged_image[:,:,1] = green_ipn_avg *5 #- 50 

ax[0, 2].imshow(merged_image)
ax[1, 2].imshow(merged_image)

height, width = red_ipn_avg.shape
merged_image = np.zeros((height, width, 3))
merged_image[:,:,0] = red_ipn_avg_post *1.5 # - 50
merged_image[:,:,1] = green_ipn_avg_post *5  #- 50 

ax[0, 3].imshow(merged_image)
ax[1, 3].imshow(merged_image)
```
