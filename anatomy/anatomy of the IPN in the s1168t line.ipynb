```python
import numpy as np
from pathlib import Path
import tifffile as tiff
import matplotlib.pyplot as plt
```

```python
ref_path = Path(r"Z:\Hagar\vision and navigation - landmarks\s11_anatomy")
anatomy = tiff.imread(ref_path / "s11_anatomy_240417_f3.tif")
```

```python
fix, ax = plt.subplots(2, 4, figsize=(8, 4))
for i in range(8):
    c = np.mod(i, 4)
    r = i // 4
    
    ax[r,c].imshow(np.rot90(anatomy[(i*2)+3], 3), cmap="gray_r", vmax=0.7)
    
    if i > 0:
        ax[r,c].axis('off')
```

```python
fix.savefig(ref_path / 'anatomy_image.pdf', dpi=300)
```
