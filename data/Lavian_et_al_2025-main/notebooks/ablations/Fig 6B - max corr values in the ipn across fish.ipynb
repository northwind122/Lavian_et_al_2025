```python
from glob import glob
import numpy as np
from scipy import stats
import flammkuchen as fl
from split_dataset import SplitDataset
from bouter import Experiment
import matplotlib.pyplot as plt

from lavian_et_al_2025.imaging.imaging_classes import TwoPExperiment
from lavian_et_al_2025.data_location import master_landmarks

from pathlib import Path
from scipy import stats
```

```python
master_ablated = master_landmarks / "2p - ablations" / 'ntr'
master_control = master_landmarks / "2p - ablations" / 'control'

fish_list_control_pre = list(master_control.glob("*_pre*"))
fish_list_ablated_pre = list(master_ablated.glob("*_pre*"))

fish_list_control_post = list(master_control.glob("*_post*"))
fish_list_ablated_post = list(master_ablated.glob("*_post*"))
```

```python
num_ntr = len(fish_list_ablated_pre)
num_control = len(fish_list_control_pre)

print(num_ntr)
print(num_control)
```

```python
count = 0

max_corr_all_ablated_pre = np.zeros((num_ntr, 354, 354))

for fish in fish_list_ablated_pre:
    suite2p_path = fish / 'suite2p'
    path_list = list(suite2p_path.glob("*00*"))
    
    num_planes = len(path_list)
    
    corr_map = fl.load(fish / 'plane_corrmap_corrvalues.h5')['plane_corr']
    max_corr = np.max(np.max(np.abs(corr_map), axis=(1)), axis=0)
    max_corr_all_ablated_pre[0,:,:] = max_corr
    
    count += 1
```

```python
count = 0

max_corr_all_ablated_post = np.zeros((num_ntr, 354, 354))

for fish in fish_list_ablated_post:
    suite2p_path = fish / 'suite2p'
    path_list = list(suite2p_path.glob("*00*"))
    
    num_planes = len(path_list)
    
    corr_map = fl.load(fish / 'plane_corrmap_corrvalues.h5')['plane_corr']
    max_corr = np.max(np.max(np.abs(corr_map), axis=(1)), axis=0)
    max_corr_all_ablated_post[0,:,:] = max_corr
    
    count += 1
```

```python
count = 0

max_corr_all_ctrl_pre = np.zeros((num_control, 354, 354))

for fish in fish_list_control_pre:
    suite2p_path = fish / 'suite2p'
    path_list = list(suite2p_path.glob("*00*"))
    
    num_planes = len(path_list)
    
    corr_map = fl.load(fish / 'plane_corrmap_corrvalues.h5')['plane_corr']
    max_corr = np.max(np.max(np.abs(corr_map), axis=(1)), axis=0)
    max_corr_all_ctrl_pre[0,:,:] = max_corr
    
    count += 1
```

```python
count = 0

max_corr_all_ctrl_post = np.zeros((num_control, 354, 354))

for fish in fish_list_control_post:
    suite2p_path = fish / 'suite2p'
    path_list = list(suite2p_path.glob("*00*"))
    
    num_planes = len(path_list)
    
    corr_map = fl.load(fish / 'plane_corrmap_corrvalues.h5')['plane_corr']
    max_corr = np.max(np.max(np.abs(corr_map), axis=(1)), axis=0)
    max_corr_all_ctrl_post[0,:,:] = max_corr
    
    count += 1
```

```python
fig, ax = plt.subplots(2, 2, figsize=(5,5))

ax[0, 0].imshow(np.rot90(np.max(max_corr_all_ablated_pre, axis=0), 3), cmap='Reds', vmin=0.05, vmax=0.5)
im = ax[0, 1].imshow(np.rot90(np.max(max_corr_all_ablated_post, axis=0), 3), cmap='Reds', vmin=0.05, vmax=0.5)

ax[1, 0].imshow(np.rot90(np.max(max_corr_all_ctrl_pre, axis=0), 3), cmap='Reds', vmin=0.05, vmax=0.5)
ax[1, 1].imshow(np.rot90(np.max(max_corr_all_ctrl_post, axis=0), 3), cmap='Reds', vmin=0.05, vmax=0.5)

ax[0, 0].axis('off')
ax[0, 1].axis('off')
ax[1, 0].axis('off')
ax[1, 1].axis('off')
```
